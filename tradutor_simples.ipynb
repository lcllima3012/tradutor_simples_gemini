```python
#importar as dependências
import os
from google.colab import userdata
from google import genai
from google.genai import types

#criando a variável de ambiente e populando com a API Key gerada no Google AI Studio
os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')

#inicializa o cliente e guarda na variável client
client = genai.Client()

#definindo o modelo a ser usado
modelo = 'gemini-2.0-flash'

#criando a configuração da resposta - como o modelo deve responder à requisição (ingles)
chat_config_ingles = types.GenerateContentConfig(
    system_instruction='Você é um poliglota fluente em inglês e irá apenas traduzir as palavras ou texto para o inglês de forma bem sucinta.',
)

#criando a configuração da resposta - como o modelo deve responder à requisição (russo)
chat_config_russo = types.GenerateContentConfig(
    system_instruction='Você é um poliglota fluente em inglês e irá apenas traduzir as palavras ou texto para o russo de forma bem sucinta.',
)

#criando a configuração da resposta - como o modelo deve responder à requisição (espanhol)
chat_config_espanhol = types.GenerateContentConfig(
    system_instruction='Você é um poliglota fluente em espanhol e irá apenas traduzir as palavras ou texto para o espanhol de forma bem sucinta.',
)

#criando a configuração da resposta - como o modelo deve responder à requisição (alemão)
chat_config_alemao = types.GenerateContentConfig(
    system_instruction='Você é um poliglota fluente em alemão e irá apenas traduzir as palavras ou texto para o alemãode forma bem sucinta.',
)

#criando a configuração da resposta - como o modelo deve responder à requisição (turco)
chat_config_turco = types.GenerateContentConfig(
    system_instruction='Você é um poliglota fluente em turco e irá apenas traduzir as palavras ou texto para o turco de forma bem sucinta.',
)

#criando a configuração da resposta - como o modelo deve responder à requisição (português)
chat_config_portugues = types.GenerateContentConfig(
    system_instruction='Você é um poliglota fluente em português do Brasil e irá apenas traduzir as palavras ou texto para o português de forma bem sucinta.',
)


#cria uma nova sessão do chat e guarda na variável chat (inglês)
chat_ingles = client.chats.create(model=modelo, config=chat_config_ingles)

#cria uma nova sessão do chat e guarda na variável chat (russo)
chat_russo = client.chats.create(model=modelo, config=chat_config_russo)

#cria uma nova sessão do chat e guarda na variável chat (alemão)
chat_alemao = client.chats.create(model=modelo, config=chat_config_alemao)

#cria uma nova sessão do chat e guarda na variável chat (espanhol)
chat_espanhol = client.chats.create(model=modelo, config=chat_config_espanhol)

#cria uma nova sessão do chat e guarda na variável chat (turco)
chat_turco = client.chats.create(model=modelo, config=chat_config_turco)

#cria uma nova sessão do chat e guarda na variável chat (pt-br)
chat_portugues = client.chats.create(model=modelo, config=chat_config_portugues)

# inicialização da variável
texto_para_traduzir = ""

# cria o loop para que o usuário insira a palavra a ser traduzida e escolha o idioma da tradução
while texto_para_traduzir != 'fim':
    texto_para_traduzir = input('Informe a palavra ou texto a ser traduzido: ')
    if texto_para_traduzir == 'fim':
        break
    else:
        try:
            opcao_lingua = int(input('Informe o idioma da tradução: \n'
                                      '1 - Ingles \n'
                                      '2 - Russo \n'
                                      '3 - Alemão \n'
                                      '4 - Espanhol \n'
                                      '5 - Turco \n'
                                      '6 - Português \n'))

            if opcao_lingua == 1:
                response = chat_ingles.send_message(texto_para_traduzir)
            elif opcao_lingua == 2:
                response = chat_russo.send_message(texto_para_traduzir)
            elif opcao_lingua == 3:
                response = chat_alemao.send_message(texto_para_traduzir)
            elif opcao_lingua == 4:
                response = chat_espanhol.send_message(texto_para_traduzir)
            elif opcao_lingua == 5:
                response = chat_turco.send_message(texto_para_traduzir)
            elif opcao_lingua == 6:
                response = chat_portugues.send_message(texto_para_traduzir)
            else:
                print('Opção inválida. Digite um número de 1 a 6')
                continue # Volta para o início do loop para pedir novamente o idioma

            print('Tradução: ', response.text)
            print('\n\n')

        except ValueError:
            print('Opção inválida. Por favor, digite um número para o idioma.')


print('Programa encerrado.')

```
